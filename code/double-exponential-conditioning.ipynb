```julia
using ApproxFun
using PyPlot
using Interact

function columns(M)
    return (view(M,:,i) for i in 1:size(M,2))
end

jac_row2(t, A1, A2, tau1, tau2) = [exp(-t/tau1), exp(-t/tau2), A1*t/tau1^2 * exp(-t/tau1), A2*t/tau2^2 * exp(-t/tau2)]
jac_row(t, A1, tau1) = [exp(-t/tau1), A1*t/tau1^2 * exp(-t/tau1)]
jac(f, args...) = hcat(f.(args...)...)'
```

```julia
function jac_neg(t, A, tau, b)
    if t < 0.0
        return [0.0, 0.0, 1.0]
    else
        return [-expm1(-t/tau), -A*t/tau^2 * exp(-t/tau), 1.0]
    end
end
```
